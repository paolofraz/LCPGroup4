```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.mixture as mix
import sklearn
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
import copy
```

```python
df = pd.read_csv("magic04.data")
df.columns = ["fLength","fWidth","fSize","fConc","fConc1","fAsym","fM3Long","fM3Trans","fAlpha","fDist","class"]
df.loc[df['class'] == 'g','class'] = 0 #gamma - signal
df.loc[df['class'] == 'h','class'] = 1 #hadron - background
```

```python
#supervised learning
ID = 1204565
cdf = copy.deepcopy(df.dropna(0, how='any'))
X = cdf.drop('class',axis=1)
Y = cdf['class']
pca = PCA(0.95, random_state=ID)
pca.fit(X)
print(np.sum(pca.explained_variance_ratio_))
print(pca.n_components_)
m = X.shape[0]
m_training = (int)(2*m/3)
m_test = m-m_training
permutation = np.random.permutation(m)
X = X.loc[permutation,:]
Y = Y[permutation]
X_train = X[0:m_training]
Y_train = Y[0:m_training]
X_test = X[m_training:]
Y_test = Y[m_training:]
```

```python
parameters = {'hidden_layer_sizes': [(10,), (50,),(100,), (150,),(200,),(250,)]}
mlp = MLPClassifier(max_iter=100, alpha=1e-4,
                    solver='sgd', tol=1e-4, random_state=ID,
                    learning_rate_init=.1)
gsCV = GridSearchCV(mlp, parameters,cv=5,return_train_score=True)
gsCV.fit(X_train,Y_train)
```
